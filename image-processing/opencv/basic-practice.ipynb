**Usig Canny edge dnetection**

```python
gray = cv2.imread(cv2.samples.findFile("baboon.jpg"), 0)

edges = []
#to compare to edged imagee
edges.append(gray)
min_val = 0

#create some patterns
while min_val <= 200 :
    #change MinVal
    edges.append(cv2.Canny(gray, min_val, 200))
    min_val += 50

row = 3
col = 2
fig, ax =  plt.subplots(nrows=row, ncols=col,  figsize=(15,15))

num = 0
fig.suptitle("Canny Edge Detection", fontsize=24, color='Black')
title = ["Original","MinVal = 0","MinVal = 50","MinVal = 100","MinVal = 150","MinVal = 200"]
for i, img in enumerate(edges):
    _r= i//col
    _c= i%col
    ax[_r,_c].set_title(title[i], fontsize=16, color='Black')
    ax[_r,_c].axes.xaxis.set_visible(False) 
    ax[_r,_c].axes.yaxis.set_visible(False) 
    ax[_r,_c].imshow(img,cmap='Greys') 
```

**Scaling**

```python
img = cv2.imread(cv2.samples.findFile("baboon.jpg"))

resized = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)

row = 1
col = 2
fig, ax =  plt.subplots(nrows=row, ncols=col,  figsize=(10,5))

fig.suptitle("Scale", fontsize=24, color='Black')

ax[0].set_title("Original", fontsize=16, color='Black')
ax[0].imshow(img)

ax[1].set_title("Scaled", fontsize=16, color='Black')
ax[1].imshow(resized)
```

```python
img = cv2.imread(cv2.samples.findFile("baboon.jpg"), 0)
rows, cols = img.shape

print('rows is {} cols is {}'.format(rows, cols))

M = np.float32([[1,0,100], [0,1,50]])
print(M)

dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst)

```

```python
img = cv2.imread(cv2.samples.findFile("baboon.jpg"), 0)
rows, cols = img.shape

M = cv2.getRotationMatrix2D((cols/2), 90, 1)
print(M)
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst)
```
